# Création de la table des abréviations

1. Importer le sondage et sa version nettoyée (2ème case)
2. Créer la table (3ème case)
3. Récupérer la table (4ème case)

```python
import numpy as np 
import pandas as pd
```

```python
#Chemin à modifier !!!!
df_clean=pd.read_csv('chemin/sondage_cleaned.csv')
df_raw=pd.read_csv('chemin/sondage2021.csv')
```

```python
colonne_raw=df_raw.columns.to_list()
colonne_abrev=df_clean.columns.to_list()
data_tuples = list(zip(colonne_raw,colonne_abrev))
df=pd.DataFrame(data_tuples, columns=['question','abreviation'])
df
```

```python
#Chemin à modifier !!!!
df.to_csv('chemin/table_abreviation.csv', index=False)
```
